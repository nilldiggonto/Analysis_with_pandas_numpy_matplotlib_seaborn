```python
############## Bar Chart ############
```

```python
soft_drinks = {'7up':30,
              'Pepsi':25,
              'Sprite':35}
```

```python
soft_drinks
```

```python
plt.figure(figsize=(12,8))
plt.bar(soft_drinks.keys(),soft_drinks.values())
plt.title('Soft Drinks')
plt.xlabel('Product Name',fontsize=12)
plt.ylabel('Price',fontsize=12)
plt.show()
```

```python
##figure
fig,ax = plt.subplots(figsize=(12,8))
ax.bar(soft_drinks.keys(),soft_drinks.values())
ax.set(title='Soft drinks',
      xlabel='Product Name',
      ylabel='Price')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.barh( list(soft_drinks.keys()), list(soft_drinks.values() )) #make a list for horizontal bar 
plt.title('Soft Drinks')
plt.xlabel('Price',fontsize=12)
plt.ylabel('Product Name',fontsize=12)
plt.show()
```
