```python
data = pd.read_html('https://github.com/mrdbourke/zero-to-mastery-ml/blob/master/data/heart-disease.csv')
```

```python
hd = data[0].iloc[:,1:]
```

```python
hd.info()
```

```python
age_over_40 = hd[hd['age']>40]
```

```python
age_over_40['age'] <40
```

```python
age_over_40['age'].min()
```

```python
age_over_40
```

```python
### Scatter using pyplot
```

```python
age_over_40.plot(x='age',y='chol',kind='scatter',c='target',cmap='rainbow',figsize=(12,8))
plt.show()
```

```python
## OOP method
```

```python
fig,ax = plt.subplots(figsize=(12,8))
age_over_40.plot(kind='scatter', x='age',y='chol',c='target',cmap='rainbow',ax=ax)
ax.set(xlim=[40,80])
plt.show()
```

```python
##########
```

```python
fig,ax = plt.subplots(figsize=(12,8))

scatter = ax.scatter(x=age_over_40['age'],y=age_over_40['chol'],c=age_over_40['target'],cmap='rainbow')

ax.set(title='Cholesterol over Age',
      xlabel='Age',
      ylabel='Cholesterol')

ax.axhline(age_over_40['chol'].mean(), linestyle='--',c='blue')

ax.legend(*scatter.legend_elements(),title='Target')

plt.show()
```

```python
#subplot of cholesterol and heart rate
```

```python
fig, (ax1,ax2) = plt.subplots(nrows= 2, ncols=1,figsize=(12,16),sharex=True)

#ax1 cholesterol
scatter = ax1.scatter(x= age_over_40['age'],y=age_over_40['chol'],c=age_over_40['target'],cmap='Spectral')

ax1.set_xlim([40,80])
ax1.set_title('Cholesterol',fontsize=20)
ax1.set(
       #xlabel='Age',
       ylabel='Cholesterol')
ax1.legend(*scatter.legend_elements(),title='Target')
#mean line
ax1.axhline(age_over_40['chol'].mean(),linestyle='--',c='red')


#ax2 heart rate
scatter_hr = ax2.scatter(x=age_over_40['age'],y=age_over_40['thalach'],c=age_over_40['target'],cmap='winter')
ax2.set_xlim([40,80])
ax2.set_title('Heart Rate',fontsize=20)
ax2.set(
       xlabel='Age',
       ylabel='Thalach')
ax2.legend(*scatter_hr.legend_elements(),title='Target')
#mean line
ax2.axhline(age_over_40['thalach'].mean(),linestyle='-.',c='blue')


fig.suptitle('Heart Disease Analysis',fontsize=25,fontweight='bold')

plt.show()
```
