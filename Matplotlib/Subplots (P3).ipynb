```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette='rainbow')
```

```python
x = np.arange(10)
```

```python
x
```

```python
y = np.random.randint(10,size=(10))
```

```python
y
```

```python
# subplot
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.scatter(x,y)

plt.subplot(2,2,3)
plt.bar(x,y)

plt.subplot(2,2,4)
plt.hist(y)

plt.show()
```

```python
fig, ( (ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2)

ax1.plot(x,y)
ax2.scatter(x,y)
ax3.bar(x,y)
ax4.hist(y)

plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(12,8))

ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[1,0].bar(x,y)
ax[1,1].hist(y)

ax[1,1].set(title='Hist',
           xlabel='X axis',
           ylabel='Y axis')

plt.show()
```
